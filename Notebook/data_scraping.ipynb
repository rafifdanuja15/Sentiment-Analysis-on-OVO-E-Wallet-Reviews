```python
from google_play_scraper import Sort, reviews
import pandas as pd
import time
```

# Data Scraping


Proses pengumpulan data pada penelitian ini dilakukan melalui teknik web scraping menggunakan library Google Play Scraper untuk mengambil ulasan pengguna aplikasi OVO di Google Play Store. Scraping dilakukan dengan parameter bahasa Indonesia (lang='id') dan lokasi Indonesia (country='id'), serta difokuskan pada ulasan terbaru (sort=Sort.NEWEST). Target pengambilan data ditetapkan sebanyak 20.000 ulasan (count=20000), di mana sistem akan berusaha mengumpulkan jumlah tersebut sesuai ketersediaan data. Hasil scraping kemudian disimpan dalam bentuk DataFrame untuk memudahkan proses analisis lebih lanjut. Dengan pendekatan ini, data yang diperoleh diharapkan merepresentasikan opini terkini pengguna terhadap aplikasi OVO.

```python
# Konfigurasi Target
app_id = 'ovo.id'
```

```python
print(f"Memulai proses scraping ulasan untuk: {app_id}")

# Melakukan scraping 20.000 ulasan terbaru
# Kita gunakan count=20000. Google Play Scraper akan berusaha mengambil sebanyak itu.
result, _ = reviews(
    app_id,
    lang='id',        # Bahasa Indonesia
    country='id',     # Lokasi Indonesia
    sort=Sort.NEWEST, # Terbaru
    count=20000       # Target jumlah ulasan
)

# Memasukkan hasil ke dalam DataFrame
df_ovo = pd.DataFrame(result)

# Menampilkan informasi dataset
print(f"Jumlah data yang berhasil diambil: {len(df_ovo)}")
```

```python
df_ovo.info()
```

# Menyimpan Data Frame

Setelah proses scraping selesai, dilakukan tahap seleksi data dengan hanya mengambil kolom yang relevan, yaitu reviewId dan content. Kolom reviewId digunakan sebagai identitas unik setiap ulasan, sedangkan content berisi teks ulasan yang akan dianalisis lebih lanjut. Data yang telah diseleksi kemudian disimpan dalam file CSV bernama ulasan_raw_ovo.csv tanpa menyertakan indeks, sehingga memudahkan proses pengolahan pada tahap berikutnya. Selain itu, dilakukan peninjauan awal dengan menampilkan beberapa data teratas untuk memastikan struktur dan isi data telah sesuai.

```python
# Memilih hanya kolom reviewId dan content
df_ovo_clean = df_ovo[['reviewId', 'content']]

# Menyimpan ke CSV mentah (hanya kolom terpilih)
df_ovo_clean.to_csv("ulasan_raw_ovo.csv", index=False)
print("File 'ulasan_raw_ovo.csv' berhasil disimpan dengan kolom reviewId dan content.")

# Menampilkan 5 data teratas untuk memastikan kolom sudah sesuai
df_ovo_clean.head()
```
